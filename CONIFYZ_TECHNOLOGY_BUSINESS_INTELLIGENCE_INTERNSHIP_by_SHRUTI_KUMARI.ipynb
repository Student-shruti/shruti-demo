#Step 1: Analyze Investment Duration Column

Locate the column in your dataset that shows investment duration, e.g.,



```python
# Step 1: Analyze Investment Duration Column
# Map categorical duration values to numerical midpoints
duration_mapping = {
    'Less than 1 year': 0.5, # Midpoint of 0 and 1 year
    '1-3 years': 2.0,      # Midpoint of 1 and 3 years
    '3-5 years': 4.0,      # Midpoint of 3 and 5 years
    'More than 5 years': 6.0   # Representative value for >5 years
}

df['Duration_Numeric'] = df['Duration'].map(duration_mapping)

# Step 2: Calculate Average Duration
average_duration = df['Duration_Numeric'].mean()

print(f"The average investment duration is: {average_duration:.2f} years")
```

#Step 2: Calculate Average Investment Duration

Once the column is clean and numeric, compute the averag
